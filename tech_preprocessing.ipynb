Этот блок применим исключительно к файлам по данной задаче. При работе повторной работе ноутбука или работе на других данных нужно будет провести оценку необходимости выполенения этого блока.

```python
df = pd.read_csv(file_name)
df_company = pd.read_csv(file_name_company)
```

```python
df = df.drop('Unnamed: 0', 1)
```

```python
df_company = df_company.set_axis(['company','city', 'product','header','review', 'mark', 'date','link'],axis=1, inplace=False)
```

```python
df = df.dropna()
df_company = df_company.dropna()
```

## 1. Техническая предобработка

### 1.1. Объединение отзыва и заголовка
Так как при отзывах информация, нужная для обработки, располагается и в заголовке, и в тексте отзыва, рациональным решением будет их объединение в одно поле

```python
df['full_review'] = df.header+' '+df.review
df_company['full_review'] = df_company.header+' '+df_company.review
```

### 1.2. Приведение даты к типу datetime
Потенциально для аналитики может быть полезной работа с датами. Приведем из дату из строки к datetime 

```python
def date_changer(df, text_field):
    df[text_field] = df[text_field].str.replace(r" янв ", "-Jan-")
    df[text_field] = df[text_field].str.replace(r" февр ", "-Feb-")    
    df[text_field] = df[text_field].str.replace(r" мар ", "-Mar-")
    df[text_field] = df[text_field].str.replace(r" апр ", "-Apr-")
    df[text_field] = df[text_field].str.replace(r" мая ", "-May-")
    df[text_field] = df[text_field].str.replace(r" июня ", "-Jun-")
    df[text_field] = df[text_field].str.replace(r" июля ", "-Jul-")
    df[text_field] = df[text_field].str.replace(r" авг ", "-Aug-")
    df[text_field] = df[text_field].str.replace(r" сент ", "-Sep-")
    df[text_field] = df[text_field].str.replace(r" окт ", "-Oct-")
    df[text_field] = df[text_field].str.replace(r" нояб ", "-Nov-")
    df[text_field] = df[text_field].str.replace(r" дек ", "-Dec-")
```

```python
date_changer(df,'date')
date_changer(df_company,'date')
```

```python
from datetime import datetime
```

```python
def change_date_csv(df,dateFormatter):
    for r in df.date:
        try:
            datetime.strptime(r, dateFormatter)
        except:
            continue
    return df
```

```python
change_date_csv(df,"%d-%b-%Y")
change_date_csv(df_company,"%d-%b-%Y")
```

### 1.3. Очистка от посторонних символов и обработка отрицаний
Для тонального анализа не представляют интерес постронние символы: служебные, или ссылки на веб-сайты, поэтому устраним эти помехи с помощью регулярных выражений

```python
def standardize_text(df, text_field):
    df[text_field] = df[text_field].str.replace(r"http\S+", "")
    df[text_field] = df[text_field].str.replace(r"http", "")
    df[text_field] = df[text_field].str.replace(r"@\S+", "")
    df[text_field] = df[text_field].str.replace(r"\sне\s", " не_")
    df[text_field] = df[text_field].str.replace(r"[^А-Яа-яA-Za-z]", " ")
    df[text_field] = df[text_field].str.lower()
    return df
```

```python
df = standardize_text(df,'full_review')
df_company = standardize_text(df_company,'full_review')
```

### 1.4. Задание тонального класса для дальнейшего обучения

```python
def det_ton(df):
    df.loc[(df.mark > 3), 'class'] = 1
    df.loc[(df.mark <= 3), 'class'] = -1
    return df
```

```python
det_ton(df)
det_ton(df_company)
```
